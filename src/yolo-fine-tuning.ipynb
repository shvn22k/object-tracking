```python
from ultralytics import YOLO
from pathlib import Path

# config
DATASET_PATH = Path("/kaggle/input/datasets/shvn22k/another-sack/sack-1")  
EPOCHS = 40
IMG_SIZE = 640
BATCH_SIZE = 8
LR = 0.0005
FREEZE = 10
MODEL_WEIGHTS = "yolo26n.pt"  



data_yaml_path = "/kaggle/working/data.yaml"

data_yaml_content = f"""
train: {DATASET_PATH}/train/images
val: {DATASET_PATH}/valid/images

nc: 1
names: ['sack']
"""

with open(data_yaml_path, "w") as f:
    f.write(data_yaml_content)

print("Data config created at:", data_yaml_path)


# load pretrained model
model = YOLO(MODEL_WEIGHTS)

# train
model.train(
    data=data_yaml_path,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    lr0=LR,
    freeze=FREEZE,
    project="/kaggle/working",
    name="sack_detection",
    exist_ok=True
)

print("Training complete.")
```
